## Word University Ranking (Data: CWUR)

#### cwurData.csv dataset (source: https://www.kaggle.com/mylesoneill/world-university-rankings)

- world_rank - world rank for university.
- institution - name of university.
- country - country of each university.
- national_rank - rank of university within its country.
- quality_of_education - rank for quality of education.
- alumni_employment - rank for alumni employment.
- quality_of_faculty - rank for quality of faculty.
- publications - rank for publications.
- influence - rank for influence.
- citations - rank for citations.
- broad_impact - rank for broad impact (only available for 2014 and 2015)
- patents - rank for patents.
- score - total score, used for determining world rank.
- year - year of ranking (2012 to 2015).

```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
```

## Creacion de Data Frame

```python
wur = pd.read_csv("data/world-university-rankings/cwurData.csv")
# vemos los primeros registros del data frame
wur.head()
```

```python
# vemos una descripcion del contenido del data frame
wur.describe()
```

```python
# filtramos valores del data frame y creamos uno nuevo solamente con informacion de argentina
wur_ar = wur[wur['country']=='Argentina']
wur_ar.head()
```

```python
# filtramos el data frame y nos quedamos solamente con informacion de USA
wur_us = wur[wur['country']=='USA']
wur_us.describe()
```

## Bokeh

Bokeh es una biblioteca de visualizacion interactiva para python que apunta a presentacion en web browsers modernos. 

Recomendamos para comenzar a crear visualizaciones visitar la [Bokeh User Guide](http://bokeh.pydata.org/en/latest/docs/user_guide.html#userguide).

Durante la clase trabajaremos con las High-level charts pero se pueden realizar visualizaciones mas complejas.


```python
from bokeh.plotting import *
from bokeh.charts import *
from bokeh.io import output_notebook

# output for jupyter
output_notebook()

from bokeh.charts import defaults

defaults.width = 500
defaults.height = 500
```

### Bar Charts

```python
# mostrar variaciones en data frame por pais y top 10
p = Bar(wur_ar, 'institution', values='broad_impact', title="Broad Impact per Year", 
        group="year", xlabel="Institucion", ylabel="Broad Impact")

show(p)
```

## Histogram

```python
# mostrar variaciones de bins
# mostrar metodo de agregacion
h = Histogram(wur, values='quality_of_education', color="green", title="US Quality of Education Rank Distribution on 2015", bins=20)
show(h)
```

```python
# mostrar separacion de por cambio de escalas del 2013 en adelante
h = Histogram(wur_us, values='quality_of_education', color='year', 
              title="US Quality of Education Rank Distribution", bins=20, xlabel="Quality of Education")
show(h)
```

## Scatter Plot

```python
s = Scatter(wur_us[wur_us['year'] > 2013], x='score', y='world_rank', title="Score vs. World Rank",
            xlabel="Score (0-100)", ylabel="World Rank", color='year', legend='top_right')

show(s)
```

```python
wur_2015 = wur[wur['year'] == 2015]
s = Scatter(wur_2015[(wur_2015['country'] == 'Argentina') | (wur_2015['country'] == 'Brazil') | (wur_2015['country'] == 'Chile') | (wur_2015['country'] == 'Uruguay')], 
            x='score', y='world_rank', title="Score vs. World Rank",
            xlabel="Score (0-100)", ylabel="World Rank", color='country', legend='top_right')
show(s)
```
